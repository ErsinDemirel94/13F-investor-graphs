```python
# Assuming your list of keys is named 'list_of_keys'
unique_interactions_cusips = interactions['base_cusip'].unique().tolist()


# Function to check if the cusip contains any of the keys as a substring
def check_keys_in_cusip(row_cusip):
    for key in unique_interactions_cusips:
        if key in str(row_cusip):
            return True
    return False

# Apply the function to the 'cusip' column to create a new 'key_in_cusip' column
merged_meta_df['found_in_interactions'] = merged_meta_df['cusip'].apply(check_keys_in_cusip)

# Display the result
merged_meta_df.head()
```

# FİLTER STOCKS

Finally, i will take subset of stocks mentioned in historical 13F filings.

By taking this subset, I filter:
    
    - stocks that has sector, industry, ticker information available in Sec api. (found in interactions)
    - stocks that has price data available in same or similar timeframe
    

```python
# Drop rows where either 'oldest' or 'newest' is NaT (Not a Time)
## filtered_df = merged_meta_df.dropna(subset=['oldest', 'newest'])

filtered_df = merged_meta_df[merged_meta_df["found_in_interactions"] == True]

# Filter rows where 'oldest' is not older than 2005 and 'newest' is not newer than 2010
#  filtered_df = filtered_df[(filtered_df['oldest'] < '2000-01-01') & (filtered_df['newest'] > '2015-12-31')]

# Display the result
filtered_df.shape
```

```python
filtered_df[filtered_df["ticker"]=="BSC1"]
```

```python
filtered_df.isna().sum()
```

```python
# Define a function to extract the base_cusip
def extract_base_cusip(cusip):
    # Split the cusip string by space and take the first part
    first_part = cusip.split(' ')[0]
    # Return the first 6 characters of the first part
    return first_part[:6]

# Apply the function to the 'cusip' column to create the 'base_cusip' column
filtered_df['base_cusip'] = filtered_df['cusip'].apply(extract_base_cusip)
```

```python
filtered_df = filtered_df.drop_duplicates(subset=['name', 'ticker'])

filtered_df.head()
```

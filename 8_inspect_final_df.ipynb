```python
import pandas as pd
from scipy import stats
import numpy as np 
# Assuming grouped_df and merged_df are defined

# Calculate percentiles for 'quarterly_volatility' in grouped_df
volatility_percentiles = aapl['quarterly_volatility'].quantile([0.05, 0.5, 0.95]).to_dict()

# Calculate percentiles for 'quarterly_volume' in grouped_df
volume_percentiles = aapl['quarterly_volume'].quantile([0.05, 0.5, 0.95]).to_dict()

def categorize_value(value):
    if pd.isna(value):
        return np.nan  # Return NaN if the value is NaN
    elif value <= volume_percentiles[0.05]:
        return 'very low'
    elif value <= volume_percentiles[0.5]:
        return 'low'
    elif value <= volume_percentiles[0.95]:
        return 'high'
    else:
        return 'very high'

# Apply the categorization functions to 'merged_df'
aapl['volatility_category'] = aapl['quarterly_volatility'].apply(categorize_value)
aapl['volume_category'] = aapl['quarterly_volume'].apply(categorize_value)
```
